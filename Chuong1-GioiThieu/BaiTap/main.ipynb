### **1. Trình bày theo ý nghĩ của riêng "bạn", tại sao bạn học lập trình?**

Em học lập trình bởi vì em rất thích việc tìm tòi và khám phá những kiến thức liên quan đến công nghệ, bên cạnh là thời đại phát triển công nghệ số nên em nghĩ sẽ có nhiều cơ hội hơn trong công việc.

### **2. Trình bày cách máy tính thực thi lệnh**

Máy tính thực thi lệnh theo chu trình 3 bước **nạp - giải mã - thực thi**.

#### 1. Giai đoạn nạp (fetch)

CPU nạp lệnh từ main memory, **Bộ đếm lệnh (Program Counter)** sẽ lưu địa chỉ của lệnh kế tiếp cần được thực thi.

Lệnh vừa được nạp sẽ được ghi vào một thanh ghi trong CPU gọi là **Thanh ghi lệnh (Instruction Register)**.

#### 2. Giai đoạn giải mã (Decode)

Sau khi lệnh được nạp vào IR, **Đơn vị điều khiển (Control Unit)** của CPU sẽ giải mã lệnh đó thành dạng dãy mã nhị phân mà máy tính có thể hiểu được.

#### 3. Giai đoạn thực thi (Execute)

Dựa trên kết quả giải mã, CPU sẽ thực thi thao tác được yêu cầu.

### **3. Trình biên dịch và trình thông dịch là gì? Lập bảng phân loại các ngôn ngữ lập trình theo trình biên dịch và trình thông dịch**

Trình biên dịch và trình thông dịch là hai loại chương trình dịch được sử dụng để chuyển đổi mã nguồn (source code) do con người viết thành mã máy (machine code) mà máy tính có thể hiểu và thực thi. Sự khác biệt giữa hai loại chương trình là cách chúng hoạt động.

#### 1. Trình biên dịch (Compiler)

Trình biên dịch dịch toàn bộ chương trình rồi mới thức thi. Nó giống như một người biên dịch sách, dịch cả cuốn sách từ ngôn ngữ này sang ngôn ngữ khác trước khi người đọc bắt đầu đọc

- Ưu điểm: Trình biên dịch chạy rất nhanh vì không cần chuyển đổi thêm bước nào trong quá trình thực thi

- Nhược điểm: Mất thời gian để dịch toàn bộ chương trình. Nếu có lỗi thì phải sửa và dịch lại từ đầu.

#### 2. Trình thông dịch (Interpreter)

Trình thông dịch dịch từng dòng chương trình một và thực thi ngay lập tức, rồi tới dòng tiếp theo. Nó giống như một người phiên dịch trực tiếp, dịch từng câu ngay khi chúng được nói ra.

- Ưu điểm: Dễ dàng phát hiện lỗi và sửa vì chúng ta có thể kiểm tra từng dòng code.

- Nhược điểm: Tốc độ thực thi chậm vì quá trình dịch diễn ra lặp đi lặp lại trong quá trình chạy chương trình.
